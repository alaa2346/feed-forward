```python
import numpy as np

def tanh(x):
    return np.tanh(x)
```

```python
np.random.seed(42)
```

```python
weights_hidden = np.random.uniform(-0.5, 0.5, (2, 2))
weights_output = np.random.uniform(-0.5, 0.5, (2, 1))
```

```python
b1 = 0.5
b2 = 0.7
```

```python
inputs = np.array([0.5, 0.3])
```

```python
hidden_layer_input = np.dot(inputs, weights_hidden) + b1
hidden_layer_output = tanh(hidden_layer_input)
```

```python
final_layer_input = np.dot(hidden_layer_output, weights_output) + b2
final_output = tanh(final_layer_input)
```

```python
print(f"Weights Hidden:\n{weights_hidden}")
print(f"Weights Output:\n{weights_output}")
print(f"Network Output: {final_output[0]}")
```
